```python
Bar plots are a type of data visualization used to represent data in the form of rectangular bars.
The height of each bar represents the value of a data point,
and the width of each bar represents the category of the data. 

# why is it used.
Bar charts are used to measure items across classes or to monitor changes over time.
These are the one of the oldest charts that represents the data visualization through the bars.
```

```python
# what is box plot ? why is it used ? plot a box plot
```

```python
A box plot is created to displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum.
In a box plot, we draw a box from the first quartile to the third quartile.
A vertical line goes through the box at the median.


# why is it used.
Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups.
They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
```

```python
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
box = [box1 , box2]
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(box)
plt.show()
```

```python
# why is suplot() function used? draw four line plots using subplot() function.
```

```python
 subplot() function adds subplot to a current figure at the specified grid position.
    It is similar to the subplots() function however unlike subplots() it adds one subplot at a time.
    So to create multiple plots you will need several lines of code with the subplot() function.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# for line 1
x = np.array([0,1,2,3,4,5])
y = np.array([0,100,200,300,400,500])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('data')
plt.show()
```

```python
# for line 2
x = np.array([0,1,2,3,4,5])
y = np.array([50,20,40,20,60,70])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('my data')
plt.show()
```

```python
# for line 3
x = np.array([0,1,2,3,4,5])
y = np.array([10,20,30,40,50,60])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('my data')
plt.show()
```
